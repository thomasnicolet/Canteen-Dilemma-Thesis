```python
## run this to make list of people changing from office to canteen, canteen to office 

# init important variables
#df = big_df
df = df1

nochange_counter = 0
canteen_to_office_counter = 0
office_to_canteen_counter = 0
canteen_to_office_list = []
office_to_canteen_list = []
no_change_list =[]

#iterate through arrival times
for i in np.unique(df['arrival']): # go through arrival times 8 - 9.1
    nochange_counter = 0
    canteen_to_office_counter = 0
    office_to_canteen_counter = 0

    df1 = df[df['arrival'] == i]
    
    #iterate through person code in new df1
    for person in np.unique(df1['code']):
    
        #set my_bool to False (0) or True (1) if the first choice they made was Office, i.e. 0 / False, or Canteen (1)
        my_bool = bool(list(df1[df1['code']==person]['choice'])[0])
        
        #skip first, then iterate through the choices that each person made, relative to the arrival time for the loop
        for choice in df1[df1['code'] == person]['choice'][1:]:
            
            # if the previos(possibly first) choice they made was office AND the next canteen:
            if (my_bool == 1 and choice == 0):
                #count the occurence, and 
                #set my_bool to 0, e.g. office
                my_bool = bool(choice) 
                canteen_to_office_counter += 1
            
            #if the previous choice was office, and next was canteen:
            elif (my_bool == 0 and choice == 1):
                #count the occurence and chance the my_bool var to 1, canteen
                office_to_canteen_counter += 1
                my_bool = bool(choice) # set my_bool to 1, e.g. canteen
                
            #this only occurs if they made the same choice  
            else:
                nochange_counter += 1
                  
    canteen_to_office_list.append(canteen_to_office_counter)
    office_to_canteen_list.append(office_to_canteen_counter)
    no_change_list.append(nochange_counter)
```

```python
# main code for generating dataset

def concat_df(csv_list):
    '''function that concats dfs'''
    my_list = []
    for i in csv_list:
        df = make_dataframe(i)
        my_list.append(df)
    df_out = pd.concat(my_list)
    return df_out

datafile = ['../Canteen_Dilemma_Data/data/Canteen_Dilemma_data/all_apps_wide_2019-02-13(batch_1).csv',
            '../Canteen_Dilemma_Data/data/Canteen_Dilemma_data/all_apps_wide_2019-02-26(batch_11).csv',
            '../Canteen_Dilemma_Data/data/Canteen_Dilemma_data/all_apps_wide_2019-02-26(batch_15).csv']
big_df = concat_df(datafile)
```
