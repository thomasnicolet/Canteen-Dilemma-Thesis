```python
from plotly.offline import init_notebook_mode, iplot
from IPython.display import display, HTML
from plotly.graph_objs import Histogram as H
import plotly.plotly as py
import pandas as pd
import numpy as np
init_notebook_mode(connected=True)
import plotly.graph_objs as go
from datetime import datetime
import seaborn as sns
from statsmodels.stats.proportion import proportion_confint
import numpy as np, scipy.stats as st


df1 = pd.read_csv('../Data/amt240_allcd.csv') # amt df
df2 = pd.read_csv('../Data/dtu1_allcd.csv') # dtu1 df
df3 = pd.read_csv('../Data/dtu2_allcd.csv') # dtu2 df
```

```python
#plot 1
#percent at canteen and percent of those being very certain, without 9:00 and 9:10
     
df = df1
#change time to be continous
df.loc[df.arrival == 9.0, 'arrival'] = 8.6
df.loc[df.arrival == 9.1, 'arrival'] = 8.7

mode_setting = 'lines+markers'
data = []
arrival_times = ["8:00","8:10", "8:20", "8:30", "8:40", "8:50", "9:00", "9:10"]
x_values = arrival_times
#orange pastel rgb(253, 205, 172)
colour1 = 'rgba(235, 99, 45, 0.9)' #orange
colour11= 'rgba(251, 220, 208, 0.6)' #light orange
#colour2 = 'rgb(224, 93, 6, 0.8)' #dark orage
#colour21 = 'rgb(253, 205, 172, 0.8)' #orange pastel light
colour2 = 'rgba(48, 116, 156, 0.9)' #light blue pastel
colour21 = 'rgba(177, 211, 231, 0.8)' #light blue
#colour2 = 'rgb(38, 123, 166, 0.5)' #dark blue
#colour21 = 'rgb(130, 195, 227, 0.5)' #dark blue

percent_canteen_list = []
temp_df = pd.DataFrame()
temp_int = 0
lower_error_bound = []
upper_error_bound = []
count = 0

for arrival in np.unique(df['arrival']):
    temp_df = df[df['arrival']== arrival]
    count = len(temp_df[temp_df['choice'] == 1])
     
    nob = len(temp_df)
    lower, upper = proportion_confint(count, nob, alpha=0.05, method='normal')
    lower_error_bound.append(lower)
    upper_error_bound.append(upper)
    percent_canteen_list.append(count / len(temp_df))
    
proper_lower_bound = []
proper_upper_bound = []
for i in range(len(percent_canteen_list)):
    proper_lower_bound.append(percent_canteen_list[i] - lower_error_bound[i])
    proper_upper_bound.append(upper_error_bound[i] - percent_canteen_list[i])

#start traces, with continuous error bars
#########################################
lower_trace1 = go.Scatter(
    opacity = 0.5,
    name='', #'Lower Bound',
    x = arrival_times,
    y = [m - n for m,n in zip(percent_canteen_list, proper_lower_bound)],
    marker=dict(
        color=colour21,
        opacity = 0.5
    ),
    line=dict(width=0),
    mode='lines'
)   
trace1 = go.Scatter(
    x = arrival_times,
    y = percent_canteen_list,
    #error_y=dict(
    #        type='data',
    #        symmetric=False,
    #        array=proper_lower_bound,
    #        arrayminus=proper_upper_bound
    #    ),
    name = 'Canteen',
    mode = mode_setting,
    line=dict(color=colour2),
    fillcolor=colour21,
    fill='tonexty'
)
upper_trace1 = go.Scatter(
    name='.95 conf.',
    opacity = 0.5,
    x = arrival_times,
    y = [m + n for m,n in zip(percent_canteen_list, proper_lower_bound)],
    mode='lines',
    marker=dict(
        color=colour21,
        opacity = 0.5
    ),#"#444"),
    line=dict(width=0),
    fillcolor=colour21,
    fill='tonexty'
)
data.append(lower_trace1)
data.append(trace1)
data.append(upper_trace1)

########################################
percent_list = []
temp_df = pd.DataFrame()
temp_int = 0
lower_error_bound = []
upper_error_bound = []
count = 0
# make a list of how many percent who go to the canteen choose 'very certain' at given time
for arrival in np.unique(df['arrival']):
    count = len(df[(df['certainty']==0.99) & (df['arrival']==arrival) & (df['choice']==1) ])
    total_count = len(df[(df['arrival']==arrival)])
    
    lower, upper = proportion_confint(count, total_count, alpha=0.05, method='normal')
    lower_error_bound.append(lower)
    upper_error_bound.append(upper)
    if total_count == 0:
        percent_list.append(0)
    else:
        percent_list.append(count / total_count)
    
    


## make traces, with continuous error bars
upper_trace2 = go.Scatter(
    name='.95 conf.',
    opacity = 0.5,
    x = x_values,
    y = upper_error_bound,
    mode='lines',
    marker=dict(
        color=colour11,
        opacity = 0.5
    ),
    line=dict(width=0),
    fillcolor=colour11,
    fill='tonexty'
)

# make data trace for percent being very certain
trace2 = go.Scatter(
    name = 'very certain',
    opacity = 0.5,
    x = x_values,
    y = percent_list,
    #error_y=dict(
    #        type='data',
    #        symmetric=False,
    #        array=proper_lower_bound,
    #        arrayminus=proper_upper_bound
    #    ),
    mode = mode_setting,
    line=dict(color=colour1),
    fillcolor=colour11,
    fill='tonexty'
)

lower_trace2 = go.Scatter(
    name= '',#'Lower Bound',
    opacity = 0.5,
    x = x_values,
    y = lower_error_bound,
    marker=dict(
        color=colour11,
        opacity = 0.5
    ),
    line=dict(width=0),
    mode='lines'
)
data.append(lower_trace2)
data.append(trace2)
data.append(upper_trace2)


##############################
#Generate annotation dictionary
key_list = [['x', 'y', 'xref', 'yref', 'text', 'showarrow', 'arrowhead', 'ax', 'ay', 'arrowsize']]*8 #*8 changed to 6 to exclude 9:0 and 9:1
dict_list = []
my_list = []
#make values
value_list = []
for i in range(0, len(arrival_times)):
    my_list = []
    my_list.append(arrival_times[i]),
    my_list.append(percent_canteen_list[i]),
    my_list.append('x'),
    my_list.append('y'),
    my_list.append(str(np.round(percent_canteen_list, 3)[i])),
    my_list.append(True),
    my_list.append(7),
    #the next two items sets position
    if i >= 8:   #only rotate annotations for 8:50
        my_list.append(-30),
        my_list.append(30)
    else:
        my_list.append(30),
        my_list.append(-30)
    my_list.append(0.75)
    value_list.append(my_list) 

dict_list = []
for i in range(0, len(key_list)):
    a = key_list[i]
    b = value_list[i]
    my_tuple = tuple(zip(a,b))
    my_dict = {}
    for key, value in my_tuple:
        my_dict[key] = value
    dict_list.append(my_dict)    
################################


#make values
value_list = []
my_list = []
for i in range(0, len(arrival_times)):
    my_list = []
    my_list.append(arrival_times[i]),
    my_list.append(percent_list[i]),
    my_list.append('x'),
    my_list.append('y'),
    my_list.append(str(np.round(percent_list, 3)[i])),
    my_list.append(True),
    my_list.append(7),
    #the next two items sets position
    if i >= 8: # rotate position 8:50
        my_list.append(40),
        my_list.append(-40)
    else:
        my_list.append(-38),
        my_list.append(38)
    my_list.append(0.75)
    value_list.append(my_list) 


for i in range(0, len(key_list)):
    a = key_list[i]
    b = value_list[i]
    my_tuple = tuple(zip(a,b))
    my_dict = {}
    for key, value in my_tuple:
        my_dict[key] = value
    dict_list.append(my_dict)
############################

#layout setting
layout = go.Layout(
    title = "[AMT] Percent going to canteen, and percent of those being very certain",
     font=dict(
        family='Courier New, monospace',
        size=14,
        color='#7f7f7f'#'#7f7f7f'
    ),
    annotations=dict_list,
    showlegend=True, #dont show legend for better use of space
    yaxis=dict(
        range = [0,1],
        title = 'Percent'
    ),
    xaxis=dict(
        title = 'Arrival times'),   
)

fig = go.Figure(data=data, layout=layout)
py.iplot(fig, filename='plot4x_percent_canteen_linechart')
#shows percentage of people going to canteen at specific times 
# filtered to whole df too
```
