```python
from plotly.offline import init_notebook_mode, iplot
from IPython.display import display, HTML
from plotly.graph_objs import Histogram as H
import plotly.plotly as py
import pandas as pd
import numpy as np
init_notebook_mode(connected=True)
import plotly.graph_objs as go
from datetime import datetime
import seaborn as sns
from statsmodels.stats.proportion import proportion_confint
import numpy as np, scipy.stats as st

df1 = pd.read_csv('../Data/amt240_allcd.csv') # amt df
df2 = pd.read_csv('../Data/dtu1_allcd.csv') # dtu1 df
df3 = pd.read_csv('../Data/dtu2_allcd.csv') # dtu2 df
```

```python
### Run for plot
from datetime import datetime
startTime = datetime.now()

#set dataframe
df = df1

canteen_coord_list = [0]*7
office_coord_list = [0]*7
disc_list = [0]*7
for session in np.unique(df['session']):

    for group in np.unique(df['group']):
        
        for round in np.unique(df[df['group']==group]['round']):
            #############################################
            office_bool = False
            canteen_bool = False
            disc_bool = False

            ### This checks whether there was discoordination or coordination in the relevant round and group
            my_bool = bool(list(df[(df['group']==group)  & (df['round']==round)]['choice'])[0])
            if (bool(list(df[(df['group']==group) & (df['round']==round)]['choice'])[0])== True) \
            & bool(list(df[(df['group']==group) & (df['round']==round)]['choice'])[1])== True:

                canteen_bool = True

            elif  (bool(list(df[(df['group']==group) & (df['round']==round)]['choice'])[0]) == False) \
                        & (bool(list(df[(df['group']==group) & (df['round']==round)]['choice'])[1]) == False):
                office_bool = True

            elif  (bool(list(df[(df['group']==group) & (df['round']==round)]['choice'])[0]) == False) \
                        & (bool(list(df[(df['group']==group) & (df['round']==round)]['choice'])[1]) == True) \
                        or ((bool(list(df[(df['group']==group) & (df['round']==round)]['choice'])[0]) == True) \
                        & (bool(list(df[(df['group']==group) & (df['round']==round)]['choice'])[1]) == False)):

                disc_bool = True

            ############################################


            ###############################################
            if ((list(df[(df['group']==group) & (df['round']==round)]['arrival'])[0] == 8.0) & \
            (list(df[(df['group']==group) & (df['round']==round)]['arrival'])[1] == 8.1)) or \
            ((list(df[(df['group']==group) & (df['round']==round)]['arrival'])[0] == 8.1) & \
            (list(df[(df['group']==group) & (df['round']==round)]['arrival'])[1] == 8.0)):

                if canteen_bool:
                    canteen_coord_list[0] +=1
                elif office_bool:
                    office_coord_list[0] +=1
                elif disc_bool:
                    disc_list[0] +=1

            elif ((list(df[(df['group']==group) & (df['round']==round)]['arrival'])[0] == 8.1) & \
            (list(df[(df['group']==group) & (df['round']==round)]['arrival'])[1] == 8.2)) or \
            ((list(df[(df['group']==group) & (df['round']==round)]['arrival'])[0] == 8.2) & \
            (list(df[(df['group']==group) & (df['round']==round)]['arrival'])[1] == 8.1)):
                if canteen_bool:
                    canteen_coord_list[1] +=1
                elif office_bool:
                    office_coord_list[1] +=1
                elif disc_bool:
                    disc_list[1] +=1

            elif ((list(df[(df['group']==group) & (df['round']==round)]['arrival'])[0] == 8.2) & \
            (list(df[(df['group']==group) & (df['round']==round)]['arrival'])[1] == 8.3)) or \
            ((list(df[(df['group']==group) & (df['round']==round)]['arrival'])[0] == 8.3) & \
            (list(df[(df['group']==group) & (df['round']==round)]['arrival'])[1] == 8.2)):
                if canteen_bool:
                    canteen_coord_list[2] +=1
                elif office_bool:
                    office_coord_list[2] +=1
                elif disc_bool:
                    disc_list[2] +=1


            elif ((list(df[(df['group']==group) & (df['round']==round)]['arrival'])[0] == 8.3) & \
            (list(df[(df['group']==group) & (df['round']==round)]['arrival'])[1] == 8.4)) or \
            ((list(df[(df['group']==group) & (df['round']==round)]['arrival'])[0] == 8.4) & \
            (list(df[(df['group']==group) & (df['round']==round)]['arrival'])[1] == 8.3)):
                if canteen_bool:
                    canteen_coord_list[3] +=1
                elif office_bool:
                    office_coord_list[3] +=1
                elif disc_bool:
                    disc_list[3] +=1

            elif ((list(df[(df['group']==group) & (df['round']==round)]['arrival'])[0] == 8.4) & \
            (list(df[(df['group']==group) & (df['round']==round)]['arrival'])[1] == 8.5)) or \
            ((list(df[(df['group']==group) & (df['round']==round)]['arrival'])[0] == 8.5) & \
            (list(df[(df['group']==group) & (df['round']==round)]['arrival'])[1] == 8.4)):
                if canteen_bool:
                    canteen_coord_list[4] +=1
                elif office_bool:
                    office_coord_list[4] +=1
                elif disc_bool:
                    disc_list[4] +=1


            elif ((list(df[(df['group']==group) & (df['round']==round)]['arrival'])[0] == 8.5) & \
            (list(df[(df['group']==group) & (df['round']==round)]['arrival'])[1] == 9.0)) or \
            ((list(df[(df['group']==group) & (df['round']==round)]['arrival'])[0] == 9.0) & \
            (list(df[(df['group']==group) & (df['round']==round)]['arrival'])[1] == 8.5)):
                if canteen_bool:
                    canteen_coord_list[5] +=1
                elif office_bool:
                    office_coord_list[5] +=1
                elif disc_bool:
                    disc_list[5] +=1

            elif ((list(df[(df['group']==group) & (df['round']==round)]['arrival'])[0] == 9) & \
            (list(df[(df['group']==group) & (df['round']==round)]['arrival'])[1] == 9.1)) or \
            ((list(df[(df['group']==group) & (df['round']==round)]['arrival'])[0] == 9.1) & \
            (list(df[(df['group']==group) & (df['round']==round)]['arrival'])[1] == 9)):
                if canteen_bool:
                    canteen_coord_list[6] +=1
                elif office_bool:
                    office_coord_list[6] +=1
                elif disc_bool:
                    disc_list[6] +=1

#print time spent
print(datetime.now() - startTime)

#######################################

# make lists above into percent plus conf intervals - necessary for plot below
# data wrangling for pot
total_list = [0]*7

for i in list(range(0, len(disc_list))):
    total_list[i] += office_coord_list[i]
    total_list[i] += canteen_coord_list[i]
    total_list[i] += disc_list[i]

#########################
percent_canteen = []
percent_office = []
percent_disc = []
upper_error_canteen = []
lower_error_canteen = []
for i in list(range(0, len(disc_list))):
    count = canteen_coord_list [i]
    total_count = total_list [i]
    lower, upper = proportion_confint(count, total_count, alpha=0.05, method='normal')
    lower_error_canteen.append(lower)
    upper_error_canteen.append(upper)
    if total_count == 0:
        percent_canteen.append(0)
    else:
        percent_canteen.append(count/total_count)
        
######################################
upper_error_office = []
lower_error_office = []
for i in list(range(0, len(disc_list))):
    count = office_coord_list[i]
    total_count = total_list [i]
    lower, upper = proportion_confint(count, total_count, alpha=0.05, method='normal')
    lower_error_office.append(lower)
    upper_error_office.append(upper)
    if total_count == 0:
        percent_office.append(0)
    else:
        percent_office.append(count/total_count)
        
######################################
upper_error_disc = []
lower_error_disc = []
for i in list(range(0, len(disc_list))):
    count = disc_list[i]
    total_count = total_list [i]
    lower, upper = proportion_confint(count, total_count, alpha=0.05, method='normal')
    lower_error_disc.append(lower)
    upper_error_disc.append(upper)
    if total_count == 0:
        percent_disc.append(0)
    else:
        percent_disc.append(count/total_count)

###############################################

#change time to be continous
df.loc[df.arrival == 9.0, 'arrival'] = 8.6
df.loc[df.arrival == 9.1, 'arrival'] = 8.7

mode_setting = 'lines+markers'
data = []
arrival_times = ["8:00/8:10", "8:10/8:20", "8:20/8:30", "8:30/8:40", "8:40/8:50", "8:50/9:00", "9:00/9:10"]
x_values = arrival_times

colour1 = 'rgba(235, 99, 45, 0.9)' #orange
colour11= 'rgba(251, 220, 208, 0.6)' #light orange
colour2 = 'rgba(48, 116, 156, 0.9)' #light blue pastel
colour21 = 'rgba(177, 211, 231, 0.8)' #light blue
c1 = 'rgba(27, 158, 119,0.9)' #green ish
c11 = 'rgba(81, 225, 182,0.6)'
c2 = 'rgba(217, 95, 2, 0.9)' # orange ish
c22 = 'rgba(253, 140, 53, 0.6)'
c3 = 'rgba(117, 112, 179,0.9)' # violet ish
c33 = 'rgba(158, 155, 202,0.6)'

#start traces, with shaded error bars
## make trace for discoordination
upper_trace1 = go.Scatter(
    name='.95 conf.',
    opacity = 0.5,
    x = x_values,
    y = upper_error_disc,
    mode='lines',
    marker=dict(
        color=c1,
        opacity = 0.5
    ),
    line=dict(width=0),
    fillcolor=c11,
    fill='tonexty'
)

# make data trace for percent being very certain
trace1 = go.Scatter(
    name = 'discoord',
    opacity = 0.5,
    x = x_values,
    y = percent_disc,
    mode = mode_setting,
    line=dict(color=c1),
    fillcolor=c11,
    fill='tonexty'
)

lower_trace1 = go.Scatter(
    name= '',#'Lower Bound',
    opacity = 0.5,
    x = x_values,
    y = lower_error_disc,
    marker=dict(
        color=c11
       # opacity = 0.5
    ),
    line=dict(width=0),
    mode='lines'
)
data.append(lower_trace1)
data.append(trace1)
data.append(upper_trace1)
########################
# trace for canteen coordination
upper_trace2 = go.Scatter(
    name='.95 conf.',
    opacity = 0.5,
    x = x_values,
    y = upper_error_canteen,
    mode='lines',
    marker=dict(
        color=c22
       # opacity = 0.5
    ),
    line=dict(width=0),
    fillcolor=c22,
    fill='tonexty'
)
trace2 = go.Scatter(
    name = 'canteen',
    opacity = 0.5,
    x = x_values,
    y = percent_canteen,
    mode = mode_setting,
    line=dict(color=c2),
    fillcolor=c22,
    fill='tonexty'
)

lower_trace2 = go.Scatter(
    name= '',#'Lower Bound',
    opacity = 0.5,
    x = x_values,
    y = lower_error_canteen,
    marker=dict(
        color=c22
        #opacity = 0.5
    ),
    line=dict(width=0),
    mode='lines'
)
data.append(lower_trace2)
data.append(trace2)
data.append(upper_trace2)
############

## make trace for office coordination
upper_trace3 = go.Scatter(
    name='.95 conf.',
    opacity = 0.5,
    x = x_values,
    y = upper_error_office,
    mode='lines',
    marker=dict(
        color=c3
       # opacity = 0.5
    ),
    line=dict(width=0),
    fillcolor=c33,
    fill='tonexty'
)

# make data trace for percent being very certain
trace3 = go.Scatter(
    name = 'office',
    opacity = 0.5,
    x = x_values,
    y = percent_office,
    mode = mode_setting,
    line=dict(color=c3),
    fillcolor=c33,
    fill='tonexty'
)

lower_trace3 = go.Scatter(
    name= '',#'Lower Bound',
    opacity = 0.5,
    x = x_values,
    y = lower_error_office,
    marker=dict(
        color=c33
        #opacity = 0.5
    ),
    line=dict(width=0),
    mode='lines'
)
data.append(lower_trace3)
data.append(trace3)
data.append(upper_trace3)

##############################
#Generate annotation dictionary
key_list = [['x', 'y', 'xref', 'yref', 'text', 'showarrow', 'arrowhead', 'ax', 'ay', 'arrowsize']]*7 #*8 changed to 6 to exclude 9:0 and 9:1
dict_list = []
my_list = []
#make values
value_list = []
for i in range(0, len(arrival_times)):
    my_list = []
    my_list.append(arrival_times[i]),
    my_list.append(percent_canteen[i]),
    my_list.append('x'),
    my_list.append('y'),
    my_list.append(str(np.round(percent_canteen, 3)[i])),
    my_list.append(True),
    my_list.append(7),
    #the next two items sets position
    if i >= 8:   #only rotate annotations for 8:50
        my_list.append(-30),
        my_list.append(30)
    else:
        my_list.append(30),
        my_list.append(-30)
    my_list.append(0.75)
    value_list.append(my_list) 

dict_list = []
for i in range(0, len(key_list)):
    a = key_list[i]
    b = value_list[i]
    my_tuple = tuple(zip(a,b))
    my_dict = {}
    for key, value in my_tuple:
        my_dict[key] = value
    dict_list.append(my_dict)    
################################

############################# annotations for office trace
value_list = []
my_list = []
for i in range(0, len(arrival_times)):
    my_list = []
    my_list.append(arrival_times[i]),
    my_list.append(percent_office[i]),
    my_list.append('x'),
    my_list.append('y'),
    my_list.append(str(np.round(percent_office, 3)[i])),
    my_list.append(True),
    my_list.append(7),
    #the next two items sets position
    if i <= 8: # rotate position 8:50
        my_list.append(0),
        my_list.append(-20)
    else:
        my_list.append(-38),
        my_list.append(38)
    my_list.append(0.75)
    value_list.append(my_list) 


for i in range(0, len(key_list)):
    a = key_list[i]
    b = value_list[i]
    my_tuple = tuple(zip(a,b))
    my_dict = {}
    for key, value in my_tuple:
        my_dict[key] = value
    dict_list.append(my_dict)
############################

#layout setting
layout = go.Layout(
    #annotations=dict_list, #not using annotations due to clutter
    title = "[AMT] Canteen and office coordination plus discoordination",
     font=dict(
        family='Courier New, monospace',
        size=14,
        color='#7f7f7f'#'#7f7f7f'
    ),
    showlegend=True, #dont show legend for better use of space
    yaxis=dict(
        range = [0,1],
        title = 'Percent'
    ),
    xaxis=dict(
        title = 'Arrival time-pair'),   
)

fig = go.Figure(data=data, layout=layout)
py.iplot(fig, filename='plot4x_percent_canteen_linechart')
#percent of canteen, office and discoordination
```
