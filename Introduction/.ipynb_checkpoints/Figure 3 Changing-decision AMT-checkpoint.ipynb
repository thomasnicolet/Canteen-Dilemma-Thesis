```python
from plotly.offline import init_notebook_mode, iplot
from IPython.display import display, HTML
from plotly.graph_objs import Histogram as H
import plotly.plotly as py
import pandas as pd
import numpy as np
init_notebook_mode(connected=True)
import plotly.graph_objs as go
from datetime import datetime
import seaborn as sns
from statsmodels.stats.proportion import proportion_confint
import numpy as np, scipy.stats as st

df1 = pd.read_csv('../Data/amt240_allcd.csv') # amt df
df2 = pd.read_csv('../Data/dtu1_allcd.csv') # dtu1 df
df3 = pd.read_csv('../Data/dtu2_allcd.csv') # dtu2 df
```

```python
## run this to make data for subsequent plots
# it makes list of people changing from office to canteen, canteen to office 
# and those who do not change even given the chance, and all those that had e.g. canteen and had the chance to change

# init important variables
df = df1

#initialize variables
nochange_counter = 0
canteen_to_office_counter = 0
office_to_canteen_counter = 0
canteen_to_office_list = []
office_to_canteen_list = []
canteen_to_canteen_list = []
office_to_office_list = []
office_x = []
canteen_x = []

#iterate through arrival times
for i in np.unique(df['arrival']): # go through arrival times 8 - 9.1
    nochange_counter = 0
    canteen_to_office_counter = 0
    office_to_canteen_counter = 0
    df1 = df[df['arrival'] == i]
    my_counter = 0
    office_to_office_counter = 0
    canteen_to_canteen_counter = 0
    canteen_counter = 0
    office_counter = 0
    
    #iterate through person codes in new df1
    for person in np.unique(df1['code']):
        #set my_bool to False (0) or True (1) if the first choice they made was Office, i.e. 0 / False, or Canteen (1)
        my_bool = bool(list(df1[df1['code']==person]['choice'])[0])
            
        #skip first, then iterate through the choices that each person made, relative to the arrival time for the loop
        for choice in df1[df1['code'] == person]['choice'][1:]:
            
            # if the previous(possibly first) choice they made was office AND the next canteen:
            if (my_bool == 1 and choice == 0):
                #count the occurence, and 
                #set my_bool to 0, e.g. office
                my_bool = bool(choice) 
                canteen_to_office_counter += 1
                canteen_counter += 1
            
            #if the previous choice was office, and next was canteen:
            elif (my_bool == 0 and choice == 1):
                #count the occurence and chance the my_bool var to 1, canteen
                office_to_canteen_counter += 1
                my_bool = bool(choice) # set my_bool to 1, e.g. canteen
                office_counter += 1
            
            #this occurs if they made "office" choice in last round, and office this round
            elif (my_bool == 0 and choice == 0):
                office_to_office_counter += 1
                
            elif (my_bool == 1 and choice == 1):
                canteen_to_canteen_counter +=1
                
            #no change counter
            #else:
            #    nochange_counter += 1
                
                  
    canteen_to_office_list.append(canteen_to_office_counter)
    office_to_canteen_list.append(office_to_canteen_counter)
    canteen_to_canteen_list.append(canteen_to_canteen_counter) # not used 
    office_to_office_list.append(office_to_office_counter) # not used
    #no_change_list.append(nochange_counter)
    #the two x lists below are total lists of those who had choose eg office and had the chance to switch
    office_x.append(office_counter+office_to_office_counter)
    canteen_x.append(canteen_counter+canteen_to_canteen_counter)
    
# Note: This counts how many made no choice at all, but does not count how many had the chance to change
```

```python
#data generation for confidence intervals for "change strategy" plots
canteen_lower_deviation = []
canteen_upper_deviation = []
for i in range(0, 8):
    
    total_count = canteen_x[i]
    count = canteen_to_office_list[i]
    lower, upper = proportion_confint(count, total_count, alpha=0.05, method='normal')
    canteen_lower_deviation.append(lower)
    canteen_upper_deviation.append(upper)
    
office_lower_deviation = []
office_upper_deviation = []
for i in range(0, 8):
    
    total_count = office_x[i]
    count = office_to_canteen_list[i]
    lower, upper = proportion_confint(count, total_count, alpha=0.05, method='normal')
    office_lower_deviation.append(lower)
    office_upper_deviation.append(upper)
```

```python
# plot_22
# run this to see barchart for each arrival_time, of people changing choice from canteen to office or vice versa
# note that this is so far the most proper normalised plot, with error bars
# data is filtered such that it counts how many percent changed their choice, out of all those who had the chance
df = df2

x_values = ['8:00', '8:10', '8:20', '8:30',  '8:40', '8:50', '9:00', '9:10'] 
#canteen
#x1 = [m / n for m,n in zip(canteen_to_office_list, canteen_x)]
x1 = []
for i in list(range(0, len(canteen_x))):
    if canteen_x[i] == 0:
        x1.append(0)
    else:
        x1.append(canteen_to_office_list[i] / canteen_x[i])

#office
#x2 = [m / n for m,n in zip(office_to_canteen_list, office_x)]
x2 = []
for i in list(range(0, len(office_x))):
    if office_x[i] == 0:
        x2.append(0)
    else:
        x2.append(office_to_canteen_list[i] / office_x[i])

#make data traces for plotly go.Figure (data) argument
#canteen_to_office
trace1 = go.Bar(
    x=['8:00', '8:10', '8:20', '8:30',  '8:40', '8:50', '9:00', '9:10'],
    y=x1,
    error_y=dict(
            type='data',
            symmetric=False,
            array=[m - n for m,n in zip(canteen_upper_deviation, x1)],
            arrayminus=[m - n for m,n in zip(x1, canteen_lower_deviation)]
        ),
    name='From canteen to office'
)

#office_to_canteen
trace2 = go.Bar(
    x=['8:00', '8:10', '8:20', '8:30',  '8:40', '8:50', '9:00', '9:10'],
    y=x2,
    error_y=dict(
            type='data',
            symmetric=False,
            array=[m - n for m,n in zip(office_upper_deviation, x2)],
            arrayminus=[m - n for m,n in zip(x2, office_lower_deviation)]
        ),
    name='From office to canteen'
)
#############################################
#Generate annotation dictionary
key_list = [['x', 'y', 'xref', 'yref', 'text', 'showarrow', 'arrowhead', 'ax', 'ay', 'arrowsize']]*len(x_values)
my_list = []
#make values
value_list = []
for i in range(0, len(x_values)):
    my_list = []
    my_list.append(x_values[i]),
    my_list.append(x1[i]), # variable for x_axis
    my_list.append('x'),
    my_list.append('y'),
    my_list.append(str(np.round(canteen_x, 2)[i])), #shown annotation
    my_list.append(True),
    my_list.append(7),
    #the next two items sets position
    my_list.append(-30),
    my_list.append(-30)
    my_list.append(0.75)
    value_list.append(my_list)
    
dict_list = []
for i in range(0, len(key_list)):
    a = key_list[i]
    b = value_list[i]
    my_tuple = tuple(zip(a,b))
    my_dict = {}
    for key, value in my_tuple:
        my_dict[key] = value
    dict_list.append(my_dict)
###############################

data = [trace1, trace2]
layout = go.Layout(
    #annotations = dict_list,
    barmode='group',
    title = "[AMT]Percent changing their decision, given the chance",
    yaxis=dict(
        range = [0,1],
        title = 'Percent'
    ),
    xaxis=dict(
        title = 'Arrival times')
)

fig = go.Figure(data=data, layout=layout)
py.iplot(fig, filename='plot22_changing_strategy_normalised_bar_chart')

#normalised percent canging decisions with error bars
#proper plot
```
